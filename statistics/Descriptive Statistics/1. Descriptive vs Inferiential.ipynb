## Statistics 
- Statistcs is a branch of mathematics where we collect and analyse data to form a conclusion.

**Type of statistics**
- Descriptive Statistics (Summary Statistics)
    - Mean
    - Madain
    - Mode
    - Standard Deviation
    - Variance
    - Sample. etc
    
- Inferential Statistics
    - Population ($\Omega$)
    - Sample ($\omega$)
     Properties of Sample:
       - Sample should be Random
       - Sample should be Representative
     - Statistics : Metric drawn on Population. Population mean( $\mu $)
     - Parameter : Metrics drawn on Sample. Sample mean ($\bar x$) 
     
     - Hypothesis testing
     - Confidence intervals. etc

## d. Cummulative Frequency

The frequency is the number of times an occurs with a given scenario. Cumulative frequency is defined as the running total frequencies. It is the sum  all previous frequencies up to the current point. It is easily understandable throught a Cumulatiuve frequency.

```python
a = df["Pclass"].value_counts()
a
```

```python
b= ((a/a.sum())*100).values
b
```

```python
c = []
cf=0
for i in b:
    cf = cf + i
    c.append(cf)
    
plt.plot(c, marker="D")    
print(c)    
```
