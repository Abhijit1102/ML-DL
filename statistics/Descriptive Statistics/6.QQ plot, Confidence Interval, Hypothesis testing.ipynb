### QQ plot
- A Q-Q (Quantile-Quantile) plot is a graphical representation of the comparison of two probability distributions. It is used to check the normality of a set of data, by comparing its distribution with a standard normal distribution. In a Q-Q plot, the quantiles of the data being analyzed are plotted against the corresponding quantiles of a standard normal distribution. If the data being analyzed follows a normal distribution, the points on the Q-Q plot should form a roughly straight line. Deviations from a straight line indicate that the data does not follow a normal distribution.

- Step in obtaining QQ plot
     - Give each point its own Quantile.
     - Give a normal curve
     - Add the same number of quantiles to the curve as you create the data.

- The Q-Q plot is a useful tool for identifying deviations from normality, which can have important implications for statistical analysis. For example, if the data is not normally distributed, certain statistical tests, such as t-tests or ANOVA, may not be appropriate and alternative tests should be used. Additionally, the Q-Q plot can also be used to identify outliers in the data and to compare the distribution of two or more sets of data.

```python
import random
a = []
for i in range(1,100):
    a.append(random.random())
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(a, dist="uniform", plot=plt)
plt.show()
```

A Q-Q graph has two axes
   - Data Quantiles (y-axis)
   - Normal Quantiles (x-axis) : -1 to 1

A Q-Q plot graph is created by plotting the quantiles of two sets of data against each other. The basic formula for constructing a Q-Q plot is as follows:

Choose a theoretical distribution, such as a normal distribution, to compare the data to.
For each data point in the sample, find the corresponding quantile in the theoretical distribution.
Plot the sample data quantile against the theoretical distribution quantile.
The resulting plot will show the sample data plotted against the theoretical distribution. If the sample data is similar to the theoretical distribution, the plotted points will form a roughly straight line.

The exact formula for finding the quantiles depends on the theoretical distribution being used. For a normal distribution, the quantiles can be calculated using the inverse cumulative distribution function (CDF) of the normal distribution.


## Confidence Intervals

Bootstrapping is a statistical method used to estimate the distribution of a sample statistic by resampling from the original sample. It is a computer-based method for estimating the sample distribution by generating multiple random samples from the original sample and computing the desired statistic for each sample.

The idea behind bootstrapping is to use the original sample as an approximation of the population and to generate additional samples from it in order to estimate the distribution of a statistic. This allows us to make inferences about the population based on the sample, even when the population distribution is unknown or difficult to estimate.

Bootstrapping is often used to estimate confidence intervals, to test hypotheses, and to perform other statistical tests. For example, if we want to estimate the 95% confidence interval for the mean of a population, we can use bootstrapping to generate multiple samples from the original sample and compute the mean for each sample. The interval that includes 95% of the computed means is then taken as the estimated 95% confidence interval for the mean.

Bootstrapping is a useful method for dealing with small or non-normal samples, and it can also be used to correct for biases or outliers in the data. However, it may not be appropriate for certain types of data or for certain types of statistical analyses, and it should be used with caution.





A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. In other words, it is an estimate of the range of values that is expected to contain the true population parameter with a certain degree of certainty.

Confidence intervals are commonly used to quantify the uncertainty in estimates of population parameters, such as the mean, median, or proportion. The level of confidence is expressed as a percentage, and it represents the degree of certainty that the true population parameter lies within the confidence interval.

For example, if a 95% confidence interval is computed for the mean of a population, this means that if the same sample were taken multiple times and a 95% confidence interval were computed for each sample, we would expect the true population mean to fall within the computed interval for 95% of the samples.

Confidence intervals are useful for making inferences about population parameters based on sample data. They allow us to take into account the variability in the sample and to estimate the range of values that is likely to contain the true population parameter. However, it is important to remember that a confidence interval is an estimate, and that the true population parameter may or may not lie within the computed interval.



## Hypothesis Testing

Hypothesis testing is a statistical procedure used to determine whether a claim or hypothesis about a population parameter is true or false. It involves formulating two opposing hypotheses (null and alternative hypothesis), selecting a sample, and collecting data. The sample data is then used to calculate a test statistic, which is compared to a critical value obtained from a probability distribution to determine whether to reject or fail to reject the null hypothesis.

The null hypothesis is typically a statement of no effect or no difference, and the alternative hypothesis is a statement of the opposite. The goal of hypothesis testing is to make a decision about which hypothesis is more supported by the data. If the test statistic falls in the critical region, we reject the null hypothesis and conclude that the alternative hypothesis is true. Otherwise, we fail to reject the null hypothesis, which means that we do not have enough evidence to support the alternative hypothesis.

Hypothesis testing is widely used in various fields, including psychology, education, economics, and many others. It helps researchers to make inferences about population parameters and to make decisions about future research or policy. However, hypothesis testing has some limitations, and it's important to choose the right statistical test and to interpret the results carefully.
