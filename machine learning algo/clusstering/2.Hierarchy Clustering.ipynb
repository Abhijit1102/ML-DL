### Clustering methods:
1. Hierarchy Clustering
2. Density Based Clustering (DBSCAN)

### Hierarchy Clustering
- Agglomerative Clustering
     - It cluster with respect to neigbourhood in hyper dimension plane and produce dentogram.
     
- Divisive Clustering
     - It will start with assumtion with one cluster and divide cluster and continue to break down.
     
**Agglomerative clustering is used widely because merging is easy w.r.t. dividing**      

### Algorithm of Agglomerative Clustering
- 1. Initiatization the proximity Matrix
- 2. Make each pint a cluster
- 3. inside a loop:
    - a. Merge two close cluster
    - b. Update the Proximity Matrix
    
- 4. Untill only one cluster is left  


### In sklearn type of Agglomerative Clustering
- 1. min(single-link) : Saseptiable to noise but helpful in cluster in a distance apart.
- 2. Max(complete link): it can control outlier and noise but if one cluster is large it will break into small cluster.
- 3. Group Average: 
       - We will compute all posible combination of each data point and average with min and max
- 4. Ward
     - it compute the centroid, and compute thier diatance and substract variance, here you minimize the variance
     
**Here you need to see the dentogram to find the cluster, by looking the vertical line which is long and vertical line sigmifies the inter cluster simmilarities**

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv("customer.csv")
```

```python
customer_data.shape
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
labels_=cluster.fit_predict(data)
labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='summer')
```
