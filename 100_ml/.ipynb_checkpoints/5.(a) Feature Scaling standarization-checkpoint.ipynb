```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds= pd.read_csv('Social_Network_Ads.csv')
```

```python
#pip install scikit-learn
```

```python
#pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
ds.head()
```

```python
df =ds.drop(["Gender", "User ID"],axis=1)
```

```python
df.columns
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(["Purchased"], axis=1),
                                                     df["Purchased"],test_size=0.3, random_state=0 )
                                                    
                                                    
```

```python
X_train.shape, X_test.shape
```

```python
#pip install StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
#scalar.fit(X_train)
scalar.fit(X_train)

X_train_scaled = scalar.transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled.shape, X_test_scaled.shape
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
## Effect of Scaling
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))

ax1.scatter(X_train["Age"], X_train["EstimatedSalary"])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled["Age"], X_train["EstimatedSalary"], color="red")
ax2.set_title("After Scaling")
plt.show()
```

```python
import seaborn as sns
```

```python
## Effect of Scaling
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))

sns.kdeplot(X_train["Age"], ax=ax1)
sns.kdeplot(X_train["EstimatedSalary"], ax=ax1)
ax1.set_title("Before Scaling")

sns.kdeplot(X_train_scaled["Age"], ax=ax2)
sns.kdeplot(X_train_scaled["EstimatedSalary"],ax=ax2)
ax2.set_title("After Scaling")
plt.show()
```

```python
## Effect of Scaling
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))

sns.kdeplot(X_train["Age"], ax=ax1)
#sns.kdeplot(X_train["EstimatedSalary"], ax=ax1)
ax1.set_title("Before Scaling")

sns.kdeplot(X_train_scaled["Age"], ax=ax2)
#sns.kdeplot(X_train_scaled["EstimatedSalary"],ax=ax2)
ax2.set_title("After Scaling")
plt.show()
```

```python
####logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred), accuracy_score(y_test, y_pred_scaled)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
dt_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
accuracy_score(y_test, y_pred), accuracy_score(y_test, y_pred_scaled)
```
