### Batch Norm ?
Batch-Notmalization is an algorthmic method which makes the training of deep Neural network faster and more stable.
It consist of normalizing activation vectors from hidden layers using the mean and variance of current batch. This normalization step is applied right before(ot after) the nonlinear function.

- normalization means mean=0, sd=1
- In batch normalization we normalized weights.

### Reason: 
- Due to non-normalization of input leads the coutor of cost function become ellipsoid leads to slow learning, but in case of normalization it became circular as hence faster convergence.
- Internal covariate shift:
      - Covarite shift: if distribution of input change but relation between output and input remain as its, but we need to changre training parameter.
      - Internal covariate shift: Its is define as the distribution of network activation due to change in netwotk parameters during training i.e. changing distribution of data in each layer while training due to updates of weights.

**Batch normalization do is ensure input of data is normalized**

- if batch normalization is not thier then leartning rate must be keep small, and low paramter initilization.

### Working:

- its used with min batch gradient desent
- it is applied layer and its optional

firstly, we can calcuate weighted function, activation function, z's are normalized -> scale shifting (gama*z_n + beta) initially (gama=1, beta=0 : learning parameter), g's(activation function) are computed, a's are computed.
     - we need to defined batch size and above process is happening and forward propogation takes place in that manner.
     
- Exponentially weighted moving average is matain during training with bach wise.

### Advantages of Batch normalization
- Stable: hyperparametric-> width range of values
- faster: learning rate higher
- regularization: helps in over-fitting not so strong, create small randomness.
- weight initilization impact reduce.


      
            




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("binary.csv")
df.head()
```

```python
tf = {"M":1,"F":0}
df.Type = df.Type.map(tf)
df.head()
```

```python
plt.scatter(df['X'],df['Y'],c=df['Type'])
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

```python
model = Sequential()

model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history1 = model.fit(X,y,epochs=200,validation_split=0.2)
```

```python
model = Sequential()

model.add(Dense(3,activation='relu',input_dim=2))
model.add(BatchNormalization())
model.add(Dense(2,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history2 = model.fit(X,y,epochs=200,validation_split=0.2)
```

```python
plt.plot(history1.history['val_accuracy'],color='black')
plt.plot(history2.history['val_accuracy'],color='green')
```
